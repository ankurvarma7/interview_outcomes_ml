```python
import pandas as pd
import numpy as np
import re
prosodic_df = pd.read_csv('/content/drive/MyDrive/prosodic_features.csv')
prosodic_df.head()
```

```python
def extract_participant_id(df):
    df['participant_id'] = df['participant&question'].str.extract(r'(P\d+)Q\d+')

    return df

prosodic_df = extract_participant_id(prosodic_df)
prosodic_df.head()
```

```python
prosodic_df['participant_numeric_id'] = prosodic_df['participant_id'].str.replace('P', '', regex=False)
prosodic_df.head()
```

```python
print(prosodic_df.columns)
```

```python
def create_interpretable_prosodic_features(df):
    grouped = df.groupby('participant_numeric_id')

    interpretable_features = []


    for participant_numeric_id, group in grouped:
        features = {
            'participant_id': participant_numeric_id,

            'average_pitch': group['mean_pitch'].mean(),
            'pitch_expressiveness': group['pitch_sd'].mean(),
            'pitch_range': group['diffPitchMaxMin'].mean(),
            'rising_intonation': group['MaxRising:3'].mean(),
            'falling_intonation': group['MaxFalling:3'].mean(),
            'speech_volume': group['intensityMean'].mean(),
            'volume_variability': group['intensitySD'].mean(),
            'volume_range': group['diffIntMaxMin'].mean(),

            'voice_steadiness': -group['jitter'].mean(),  # Negative because lower jitter is better
            'volume_steadiness': -group['shimmer'].mean(),  # Negative because lower shimmer is better
            'voice_break_frequency': group['numVoiceBreaks'].mean(),

            'speaking_pace': group['speakRate'].mean(),
            'pause_frequency': group['numPause'].sum() / group['duration'].sum(),  # Pauses per second
            'average_pause_length': group['avgDurPause'].mean(),
            'longest_pause': group['maxDurPause'].mean(),
            'average_response_length': group['duration'].mean()
        }

        interpretable_features.append(features)

    return pd.DataFrame(interpretable_features)

interpretable_prosodic_df = create_interpretable_prosodic_features(prosodic_df)
```

```python
scores_df = pd.read_csv('/content/drive/MyDrive/scores.csv')
scores_df.head()

scores_df['participant_numeric_id'] = scores_df['Participant'].str.replace('p', '', regex=False)
scores_df.head()

grouped_scores = scores_df.groupby('participant_numeric_id')

average_scores = []
for participant_numeric_id, group in grouped_scores:
    features = {
        'participant_id': participant_numeric_id,
        'overall_score_mean' : group['Overall'].mean(),
        'excitement_score_mean' : group['Excited'].mean(),
    }
    average_scores.append(features)

average_scores_df = pd.DataFrame(average_scores)
print(average_scores_df)
```
